```python
x = np.arange(0,10*np.pi,0.1)
y = np.sin(x) 
z = np.cos(x)
fig, ax = plt.subplots()
ax.plot(x,z,color='blue')
ax.fill_between(x,z,color='blue',alpha = 0.3)
plt.show()
```

```python
#LINE GRAPH 
a = [1,2,3,4,5,6] 
b = [5,4,6,2,3,4]
c = (1,4,8,3,5,2)

#Assign a background color to the plot 
ax = plt.axes() 
ax.set_facecolor("pink")

plt.title("LINE GRAPH")
plt.plot(a,b,color='red',label = 'LINE 1',linewidth = 5)
plt.plot(a,c,color='blue',label = 'LINE 2',linestyle = 'dashdot',marker = "^",markersize = 10,markerfacecolor='green',markeredgecolor='green' )
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.legend()
plt.show()
```

```python
# BAR GRAPH  
x = ["AAA","BBB","CCC","DDD","EEE"]
y = [50,15,37,25,63]
plt.title("SIMPLE BAR GRAPH",fontsize = 15)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xticks(rotation = 45)

color12 = ['red','green','black','orange','pink']
plt.bar(x,y,color=color12)
plt.grid()
plt.savefig("bar_graph.png")#saves your image
plt.show()
```

```python
# PIE CHART 
#composition of air 

# plt.figure(figsize=(15,15))
values = ['57','21','10','13','4']
x = ['Nitrogen','oxygen','co2','hydrogen','other gases']
c2 = ['red','g','y','#FFFE00','#E700FF']
exp1 = [0,0,0.2,0,0]
plt.pie(values,labels=x,shadow = True,autopct = '%1.1f%%',explode = exp1,colors = c2,counterclock=False,startangle = 90)
plt.legend(loc = 'lower right')
plt.show()
```

```python
#Histogram 
x = np.random.normal(10,15,250) #normal distribution 
plt.hist(x,histtype = 'stepfilled',color = 'red')
# 'bar', 'barstacked', 'step', 'stepfilled
plt.show()
```

```python
#SCATTER PLOT
a = [1,2,3,4,5,6,7,8]
b = [5,4,7,3,2,9,1,3]
c12 = ['r','g','b','c','m','y','k','purple']
plt.scatter(a,b,marker='^',color = c12,s=100,alpha = 0.6)
```
