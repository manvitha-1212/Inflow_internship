```python
def os_category(input1):
  if input1 =='Windows 10' or input1 == 'Windows 7' or input1 == 'Windows 10 S':
    return 'Windows'
  elif input1 == 'macOS' or input1 == 'Mac OS X':
    return 'Mac'
  else:
    return 'Linux/others/No OS'
```

```python
df2["operating system"]=df2["OpSys"].apply(os_category)
```
